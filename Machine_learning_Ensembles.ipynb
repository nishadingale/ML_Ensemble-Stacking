## Advanced Techniques:
* Bagging (Bootstrap AGGregatING)
* Boosting

```
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier

wine = load_wine()

X,y = wine.data[:,11:13],wine.target
```

```
clf = DecisionTreeClassifier()
clf.fit(X,y)
```

```
clf.score(X,y)
```

```
from sklearn.ensemble import BaggingClassifier

bg = BaggingClassifier(base_estimator=clf,max_features=0.1,max_samples=0.1,random_state=1234)
```

```
bg.fit(X,y)
```

```
bg.score(X,y)
```

```
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X,y,bg)
```

# Task Three

```
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion = "entropy",max_depth = 1)
print(clf)
clf.fit(X,y)
print(clf.score(X,y))
plot_decision_regions(X,y,clf)
```

```
from sklearn.ensemble import AdaBoostClassifier
import matplotlib.pyplot as plt 

clf = DecisionTreeClassifier(criterion = "entropy",max_depth = 1)

n_est = [1,2,5,10,100]
for i in n_est:
  clf_boost = AdaBoostClassifier(clf,n_estimators = i)
  print(i)
  clf_boost.fit(X,y)
  print(clf_boost.score(X,y))
  plot_decision_regions(X,y,clf_boost)
  plt.show()
```

# Task Four

## Boosting algorithms:

* AdaBoost : Adaptive boosting
* GBM : Gradient Boosting
* XGBM : Extreme Gradient Boosting
* Light GBM
* CatBoost

```
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
import xgboost as xgb
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)



learning_rate = [0.001,0.01,0.1,1,5,10]
for i in learning_rate:
  model = GradientBoostingClassifier(random_state=1234,learning_rate= i)
  print("learning rate:",i)
  model.fit(X_train,y_train)
  print("score :",model.score(X_test,y_test))
  plot_decision_regions(X_train,y_train,model)
  plt.show()
```

```

learning_rate = [0.001,0.01,0.1,1,5,10]
for i in learning_rate:
  model = xgb.XGBClassifier (random_state=1234,learning_rate= i)
  print("learning rate:",i)
  model.fit(X_train,y_train)
  print("score :",model.score(X_test,y_test))
  plot_decision_regions(X_train,y_train,model)
  plt.show()
```

# Task Five

## Ensemble Learning:
* Homogeneous ensemble methods : Bagging , Boosting
* Heterogeneous ensemble methods: Stacking

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,StackingClassifier

```

```

estimators = [("clfk",KNeighborsClassifier(n_neighbors = 1)),("clfg",GaussianNB()),("clfr",RandomForestClassifier())]
final_estimator = LogisticRegression()


st_clf = StackingClassifier(estimators= estimators,final_estimator=final_estimator)
```

```
clfs = [clfk,clfg,clfr,st_clf]

for classifier in clfs:
  classifier.fit(X_train,y_train)
  print(f"Score for {classifier}:",classifier.score(X_test,y_test))
  plot_decision_regions(X_test,y_test,classifier)
  plt.show()
```
